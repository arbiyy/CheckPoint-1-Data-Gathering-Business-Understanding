### **Outliers Values**

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

Akan tetapi memang tidak semua kolom perlu kita tangani, kita harus tetap melihat dan memahami tujuan dari setiap kolom yang ada di dataset.

Dan terlihat kolom
* Year

Dengan presentase 0.7% (dibawah 1%)
* Kilometers_driven

Dengan presentase 3%

* Seats

Dengan presentase 0.6%

* Price

Dengan presentase 9%

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

Aktivitas: Membandingkan total **harga jual (Price)** di setiap **lokasi (Location)**.

Tujuan: Mengidentifikasi lokasi mana yang memiliki total **nilai transaksi terbesar** dan membandingkan **distribusi nilai mobil antar lokasi** secara langsung.

Visualisasi: Bar Chart (Grafik Batang).

```python
penjualan_location = df.groupby('Location')['Price'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=penjualan_location.index, y=penjualan_location.values, palette='viridis', hue=penjualan_location.index, legend=False)
plt.title('Total Price by Location')
plt.xlabel('Location')
plt.ylabel('Total Price')
plt.xticks(rotation=45)
plt.show()
```

## **Composition/Komposisi**

Aktivitas: Melihat proporsi atau persentase total **harga jual (Price)** berdasarkan **tipe bahan bakar (Fuel_Type)**.

Tujuan: Mengetahui tipe bahan bakar mana yang paling mendominasi portofolio penjualan kendaraan secara keseluruhan berdasarkan total harga jual.

Visualisasi: Pie Chart (Grafik Lingkaran).

```python
price_by_fuel_type = df.groupby('Fuel_Type')['Price'].sum().sort_values(ascending=False)
price_by_fuel_type_top3 = price_by_fuel_type.head(3)

plt.figure(figsize=(10, 8))
price_by_fuel_type_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Tipe Bahan Bakar Teratas Berdasarkan Total Harga Jual')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## Heatmap

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data=df[['Year', 'Kilometers_Driven', 'Seats', 'Price']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Fitur Numerik Penting', fontsize=16)
plt.show()
```

Grafik ini berjudul 'Korelasi antara Fitur Numerik Penting'.

  -Setiap kotak dalam heatmap menunjukkan nilai korelasi antara dua fitur numerik.

  -annot=True menampilkan nilai korelasi numerik di setiap kotak.

  -cmap='viridis' mengatur skema warna, di mana warna yang lebih terang atau gelap menunjukkan kekuatan korelasi yang berbeda (positif atau negatif).

  -fmt='.2f' memformat nilai korelasi menjadi dua angka di belakang koma.

Dari heatmap ini, Anda bisa melihat:

  -Korelasi Positif (nilai mendekati 1): Semakin tinggi nilai satu fitur, semakin tinggi pula nilai fitur lainnya (misalnya, Price dan Year mungkin menunjukkan korelasi positif jika mobil baru cenderung lebih mahal).

  -Korelasi Negatif (nilai mendekati -1): Semakin tinggi nilai satu fitur, semakin rendah nilai fitur lainnya (misalnya, Price dan Kilometers_Driven cenderung memiliki korelasi negatif, artinya semakin banyak jarak tempuh, semakin rendah harganya).
  
  -Tidak Ada Korelasi (nilai mendekati 0): Tidak ada hubungan linear yang jelas antara dua fitur.
