### Bussiness Understanding

Pada tahap ini, kita melihat konteks industri otomotif. Dataset ini berisi informasi penting seperti merek, tahun, harga jual, harga baru, jarak tempuh (km), jenis bahan bakar, tipe penjual, transmisi, dan jumlah pemilik sebelumnya.

1. Masalah: Penentuan harga mobil bekas sangat subjektif. Dealer atau platform jual-beli sering kesulitan menentukan harga yang "pas"—tidak terlalu mahal agar cepat laku, tapi tidak terlalu murah agar tetap untung.

2. Konteks: Fluktuasi harga mobil bekas sangat dipengaruhi oleh depresiasi (penurunan nilai) berdasarkan usia dan pemakaian (jarak tempuh).
  

### Bussiness Objective

Deskripsi Dataset:
Dataset ini merupakan representasi pasar mobil bekas yang mencakup variabel kritis seperti Selling Price (harga jual saat ini), Present Price (harga baru dari pabrikan), Kms Driven (jarak tempuh), Fuel Type, Transmission, dan Owner (jumlah tangan sebelumnya). Data ini mencerminkan dinamika depresiasi kendaraan di pasar otomotif.

Tujuan Analisis Bisnis:

1. Optimalisasi Strategi Penetapan Harga (Dynamic Pricing): Membangun sistem estimasi harga otomatis agar dealer dapat memberikan penawaran beli/jual yang kompetitif namun tetap menjaga margin keuntungan.

2. Analisis Retensi Nilai (Asset Depreciation): Mengidentifikasi karakteristik mobil (seperti jenis bahan bakar atau transmisi) yang memiliki nilai jual kembali (resale value) paling stabil untuk dijadikan prioritas stok inventori.

3. Segmentasi Stok Berdasarkan Efisiensi: Menentukan kombinasi jarak tempuh dan usia kendaraan yang paling diminati pasar guna mempercepat perputaran unit (inventory turnover) dan mengurangi biaya penyimpanan di gudang.

4. Personalisasi Penawaran Pelanggan: Memberikan rekomendasi harga yang transparan kepada pelanggan berdasarkan perbandingan harga pasar, guna meningkatkan kepercayaan dan loyalitas konsumen.
